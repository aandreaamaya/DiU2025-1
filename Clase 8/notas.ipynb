# Filtros de Variables
### Ejemplos de filtros para el proyecto:
**date: Formato de Fechas **

< p >Fecha de creación: {{ habito.fecha_creacion|date:"d/m/Y" }}< /p >

"d/m/Y": Este formato muestra la fecha como día/mes/año (por ejemplo, 12/10/2024).

**time: Formato de Hora**

< p >Hora del recordatorio: {{ habito.hora_recordatorio|time:"H:i" }}< /p >

"H:i": Muestra la hora en formato de 24 horas (por ejemplo, 14:30).

**timesince: Tiempo Transcurrido**

< p >Hace {{ habito.fecha_inicio|timesince }} que empezaste este hábito.< /p >


Ejemplo: "Hace 2 semanas".

**pluralize: Pluralización Automática**

< p >Has completado {{ habito.repeticiones }} repetición{{ habito.repeticiones|pluralize }} de este hábito.< /p >

Si el valor de repeticiones es 1, mostrará "repetición".

Si el valor de repeticiones es mayor que 1, mostrará "repeticiones".

**yesno: Mostrar Texto Condicional**

< p >Este hábito ha sido completado: {{ habito.completado|yesno:"Sí,No" }}< /p >

Si el valor es verdadero, mostrará "Sí", de lo contrario "No".

**floatformat: Formato Decimal**

< p >Progreso: {{ habito.progreso|floatformat:1 }}%< /p >

floatformat:1: Muestra el número con un decimal (por ejemplo, 75.3%).

**default: Valor Predeterminado**

< p >Recordatorio: {{ habito.recordatorio|default:"Sin recordatorio" }}< /p >

Si el valor de recordatorio es None, mostrará "Sin recordatorio".

**length: Longitud de una Lista**

< p >Tienes {{ lista_habitos|length }} hábitos.< /p >

**join: Unir Elementos de una Lista**

< p >Categorías: {{ habito.categorias|join:", " }}< /p >

Si categorias es una lista de categorías, mostrará: "Categoría1, Categoría2, Categoría3".

**truncatechars y truncatewords: Truncar Texto**

< p >Descripción: {{ habito.descripcion|truncatewords:10 }}< /p >

truncatewords:10: Trunca el texto después de 10 palabras.

**slice: Cortar una Lista**
< ul >
    {% for habito in lista_habitos|slice:":5" %}
        < li >{{ habito.nombre }}< /li > 
    {% endfor %}
< /ul > 

slice:":5": Muestra solo los primeros 5 hábitos de la lista.

**dictsort: Ordenar Listas de Diccionarios**

< ul >
    {% for habito in lista_habitos|dictsort:"fecha_inicio" %}
        < li >{{ habito.nombre }} - Fecha: {{ habito.fecha_inicio|date:"d/m/Y" }}< /li >
    {% endfor %}
< /ul >

dictsort:"fecha_inicio": Ordena los hábitos por la clave fecha_inicio.

**capfirst: Capitalizar la Primera Letra**

< p >{{ habito.nombre|capfirst }}< /p >

**linebreaksbr: Convertir Saltos de Línea en Etiquetas < br >**

< p >{{ habito.descripcion|linebreaksbr }}< /p >

**wordcount: Contar Palabras**

< p >Este hábito tiene {{ habito.descripcion|wordcount }} palabras en su descripción.< /p >

#% csrf\_token \%
### < form method="post" >
###    {% csrf_token %}
###    < button type="submit" > Enviar < /button >
### < /form >

```python
function getCookie(name) {
    let cookieValue = null;
    if (document.cookie && document.cookie !== '') {
        const cookies = document.cookie.split(';');
        for (let i = 0; i < cookies.length; i++) {
            const cookie = cookies[i].trim();
            if (cookie.substring(0, name.length + 1) === (name + '=')) {
                cookieValue = decodeURIComponent(cookie.substring(name.length + 1));
                break;
            }
        }
    }
    return cookieValue;
}

const csrftoken = getCookie('csrftoken');

fetch('/submit-form/', {
    method: 'POST',
    headers: {
        'X-CSRFToken': csrftoken,
        'Content-Type': 'application/json',
    },
    body: JSON.stringify({ name: 'John' })
})
.then(response => response.json())
.then(data => console.log(data));
```
